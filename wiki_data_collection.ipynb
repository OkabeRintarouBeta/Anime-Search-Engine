# filter out all rows that are not English

```python
def isEnglish(s):
    try:
        s.isascii()
    except UnicodeDecodeError:
        return False
    else:
        return True

isEnglish_list=[]
anime_name_list=[]

for idx,row in df_animes_filtered.iterrows():
    wiki_url=row['wiki_link']
    
    wiki_url=unquote(wiki_url)
#     print(wiki_url)
    try:
        anime_name=re.findall('https*://en.wikipedia.org/wiki/(.*)', wiki_url)[0]
        anime_name_list.append(anime_name)
#         print(anime_name)
        isEnglish_list.append(isEnglish(anime_name))
    except:
        isEnglish_list.append(False)
        anime_name_list.append("")
    
        
df_animes_filtered['isEnglish']=isEnglish_list
df_animes_filtered['anime_name']=anime_name_list

df_animes_filtered=df_animes_filtered[df_animes_filtered['isEnglish']==True]
df_animes_filtered.drop(columns=['isEnglish'],inplace=True)
df_animes_filtered.to_csv("animes_filtered.csv")
```

# extract wiki intro

```python
import wikipediaapi
from urllib.parse import unquote
import re

wiki_wiki = wikipediaapi.Wikipedia('en')

# page_py = wiki_wiki.page('Haikyu!!')
# print("Page - Exists: %s" % page_py.exists())


# url='http://en.wikipedia.org/wiki/Haiky%C5%AB!!'
def extract_intro(index,anime_name, level=0):
    page_py = wiki_wiki.page(anime_name)
#     print("Page - Exists: %s" % page_py.exists())
    sections=page_py.sections
    flag=False
    intro=""
    possible_sections=['Setting','Settings','Plot','Plots','Background','Synopsis','Story','Plots and settings','Plot and setting','Plots and setting','Plot and settings']
#     print("sections:",sections)
    for s in sections:
#         print("=================================")
#         print(s.title)
        if s.title in possible_sections:
            flag=True
#             print(s.text)
            intro+=s.text
            intro+="\n"
#             print(intro)
#         print("%s: %s - %s" % ("*" * (level + 1), s.title, s.text[0:40]))
#         print("***********************************")
        for sub in s.sections:
            if sub.title in possible_sections:
                flag=True
                intro+=sub.text
                intro+="\n"
    if not flag:
        print(index,":",anime_name+" does not have plot intro")
    if intro=='':
        intro="none"
    return intro

```
